```python
# This is similar to [Pulse_classification_largesize.ipynb] except
#  that data processing is wrapped in customized functions
# %matplotlib notebook
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the history data
with open('../data/model/tcn/comparison/history_tcn_v0.h5', 'rb') as f:
    history_tcn= pickle.load(f)
with open('../data/model/lstm/comparison/history_bilstm_v0.h5', 'rb') as f:
    history_bilstm= pickle.load(f)
```

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 10})
matplotlib.rcParams["font.family"] = "Arial"
matplotlib.rcParams.update({'axes.linewidth': 2})

fig, ax= plt.subplots(1,2, figsize=(10,3.5), dpi=300)
linewidth=3
ax[0].plot(history_tcn["loss"], linewidth=2, 
           color='tab:green', label="HD-CNN loss")
ax[0].plot(history_tcn["val_loss"], linestyle='dashed', linewidth=linewidth, 
           color='tab:purple', label="HD-CNN val_loss")
ax[0].plot(history_bilstm["loss"], linewidth=2, 
           color='tab:orange', label="Bi-LSTM loss")
ax[0].plot(history_bilstm["val_loss"], linestyle='dashed', linewidth=linewidth,
           color='tab:blue', label="Bi-LSTM val_loss")
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss')
ax[0].legend(bbox_to_anchor=(0.9, 0.98))

ax[1].plot(history_tcn["accuracy"], linewidth=2, 
           color='tab:green', label="HD-CNN acc")
ax[1].plot(history_tcn["val_accuracy"], linestyle='dashed', linewidth=linewidth, 
           color='tab:purple', label="HD-CNN val_acc")
ax[1].plot(history_bilstm["accuracy"], linewidth=2, 
           color='tab:orange', label="Bi-LSTM acc")
ax[1].plot(history_bilstm["val_accuracy"], linestyle='dashed', linewidth=linewidth,
           color='tab:blue', label="Bi-LSTM val_acc")
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Accuracy')
ax[1].set_ylim([0, 1])
ax[1].legend(bbox_to_anchor=(0.9, 0.4))
```

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 10})
matplotlib.rcParams["font.family"] = "Arial"
matplotlib.rcParams.update({'axes.linewidth': 2})

fig, ax= plt.subplots(figsize=(2,0.7), dpi=300)
linewidth=3
ax.plot(history_tcn["loss"], linewidth=2, 
           color='tab:green', label="1D-DCNN loss")
ax.plot(history_tcn["val_loss"], linestyle='dashed', linewidth=linewidth, 
           color='tab:purple', label="1D-DCNN val_loss")
ax.plot(history_bilstm["loss"], linewidth=2, 
           color='tab:orange', label="Bi-LSTM loss")
ax.plot(history_bilstm["val_loss"], linestyle='dashed', linewidth=linewidth,
           color='tab:blue', label="Bi-LSTM val_loss")
ax.set_xlim([40, 100])
ax.set_ylim([0, 0.2])
```

```python
fig, ax= plt.subplots(figsize=(2,0.7), dpi=300)

ax.plot(history_tcn["accuracy"], linewidth=2, 
           color='tab:green', label="1D-DCNN acc")
ax.plot(history_tcn["val_accuracy"], linestyle='dashed', linewidth=linewidth, 
           color='tab:purple', label="1D-DCNN val_acc")
ax.plot(history_bilstm["accuracy"], linewidth=2, 
           color='tab:orange', label="Bi-LSTM acc")
ax.plot(history_bilstm["val_accuracy"], linestyle='dashed', linewidth=linewidth,
           color='tab:blue', label="Bi-LSTM val_acc")
ax.set_xlim([40, 100])
ax.set_ylim([0.96, 1.0])
```
